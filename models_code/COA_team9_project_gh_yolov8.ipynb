### Load a Pre-trained Model

```python
from ultralytics import YOLO

model = YOLO('yolov8m.pt')
```

```python
print(type(model.names), len(model.names))

print(model.names)
```

```python
model.train(data='./foods-4/food_data.yaml', epochs=50, patience=30, batch=32, imgsz=640, device=0)
```
